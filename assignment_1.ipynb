Question 3:

There are several data types in Python Programming language. The three of which are:
1. String (str): String is a data type which can comprise of any kind of character (be it alphanumeric or any other kind of character) but it has to enclosed in single quotes ('') or double quotes (""). Strings are immutable which means that it can't be mutated or more technically, it's identity (id) remains the same throughout its life in the program. If we run our program again and agin, then it may show different different id's everytime. If we try to change the contents of a string variable, it's 'id'
will change which shows us that a new variable (by the same name) has been created and the old one has been erased. It's not like we mutated the original string. Example: 'abcdef' or 'are457' or '123'

2. Integer (int): 'int' or integer is a data type that is of numeric type and it has no decimal value. 'Int' data type is also immutable which means if try to mutate it, a new variable will be created with a different id.
Example: 1 or 4 or -5.

3. Float: Float is a data type that is of numerical type but it has some decimal value. 'float' data type is also immutable which means if try to mutate it, a new variable will be created with a different id.
Example: 1.5 or 4.0 or -3.2.

2) To get a floating point number version of a value, we can use float() function. If we try to pass any other kind of data type to float() function, then the code will show error.

Question 10:


'I have eaten' + 99 + 'burritos'
This expression is causing an error because we are trying to concatenate two string values ('I have eaten' & 'burritos') to a integer value (which are two different data types) which is not possible in Python.


What we can do to fix it is that we can typecast the int value 99 to str type by using str() function so that all the three values are now homogeneous.


'I have eaten' + str(99) + 'burritos'
